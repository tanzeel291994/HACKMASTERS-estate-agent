```python
from py2neo import Graph
import pandas as pd
from typing import Dict, List, Optional
import ast
import json
from openai import AzureOpenAI


class RealEstateKnowledgeGraph:
    
    def __init__(self):
        import os
        self.graph = Graph("bolt://localhost:7687", auth=("neo4j", "1234neo4j"))
        self.embedding_model = AzureOpenAI(
            api_version="2024-12-01-preview",
            azure_endpoint="https://bodeu-des-csv02.openai.azure.com/",
            api_key=os.getenv("AZURE_OPENAI_KEY")
        )
        self.embedding_model_name = "text-embedding-3-small"
    
    def get_embedding(self, text):
        response = self.embedding_model.embeddings.create(
            input=[text],
            model=self.embedding_model_name
        )
        return response.data[0].embedding
    
    def create_knowledge_graph(self, df_amenities: pd.DataFrame):
        # Create constraints for unique nodes
        self.graph.run("MATCH (n) DETACH DELETE n")
        #'transport','metro_stations', 'gyms', 'spas', 'supermarkets', 'malls', 'schools','restaurant', 'hospital', 'tourist_attraction', 'park', 'worship'
        self.graph.run("CREATE CONSTRAINT IF NOT EXISTS FOR (a:Area) REQUIRE a.name IS UNIQUE")
        self.graph.run("CREATE CONSTRAINT IF NOT EXISTS FOR (b:Building) REQUIRE b.name IS UNIQUE")
        self.graph.run("CREATE CONSTRAINT IF NOT EXISTS FOR (s:School) REQUIRE (s.name, s.vicinity) IS UNIQUE")
        self.graph.run("CREATE CONSTRAINT IF NOT EXISTS FOR (m:MetroStation) REQUIRE (m.name, m.vicinity) IS UNIQUE")
        self.graph.run("CREATE CONSTRAINT IF NOT EXISTS FOR (h:Hospital) REQUIRE (h.name, h.vicinity) IS UNIQUE")
        self.graph.run("CREATE CONSTRAINT IF NOT EXISTS FOR (t:Transport) REQUIRE (t.name, t.vicinity) IS UNIQUE")
        self.graph.run("CREATE CONSTRAINT IF NOT EXISTS FOR (g:Gym) REQUIRE (g.name, g.vicinity) IS UNIQUE")
        self.graph.run("CREATE CONSTRAINT IF NOT EXISTS FOR (s:SPA) REQUIRE (s.name, s.vicinity) IS UNIQUE")
        self.graph.run("CREATE CONSTRAINT IF NOT EXISTS FOR (s:Supermarket) REQUIRE (s.name, s.vicinity) IS UNIQUE")
        self.graph.run("CREATE CONSTRAINT IF NOT EXISTS FOR (m:Mall) REQUIRE (m.name, m.vicinity) IS UNIQUE")
        self.graph.run("CREATE CONSTRAINT IF NOT EXISTS FOR (r:Restaurant) REQUIRE (r.name, r.vicinity) IS UNIQUE")
        self.graph.run("CREATE CONSTRAINT IF NOT EXISTS FOR (t:TouristAttraction) REQUIRE (t.name, t.vicinity) IS UNIQUE")
        self.graph.run("CREATE CONSTRAINT IF NOT EXISTS FOR (p:Park) REQUIRE (p.name, p.vicinity) IS UNIQUE")
        self.graph.run("CREATE CONSTRAINT IF NOT EXISTS FOR (w:Worship) REQUIRE (w.name, w.vicinity) IS UNIQUE")


        # First, create all areas
        areas = df_areas_info['location_y'].unique()
        for area_name in areas:
            print(area_name)
            row = df_areas_info[df_areas_info["location_y"] == area_name].iloc[0]
            self._create_area(row)

        # Then create buildings and their relationships
        for _, row in df_buildings_info.iterrows():
            row_amenties_info = df_amneties[df_amneties["building_name"] == row["building_name"]].iloc[0]
            self._create_building_with_amenities(row_amenties_info,row)

        # ... similar for other amenities ...

    def _create_area(self,row):
        area_cols = ['about-', 'in-a-nutshell', 'community-overview',
       'properties-in', 'transportation-and-parking-spaces',
       'supermarkets-near', 'mosques-near', 'other-places-of-worship',
       'schools-near', 'clinics-and-hospitals-near', 'nearby-areas',
       'malls-near', 'restaurants-near', 'beaches-near',
       'leisure-activities-and-notable-landmarks',
       'outdoor-activities,-fitness-and-beauty', 'community-events',
       'social-community-groups', 'things-to-consider', 'location_x', 'area',
       'img_name', 'location_y', 'sub_location', 'status']
        
        area_details = []
        for col in area_cols:
            if col in row and pd.notna(row[col]) and row[col]:
                area_details.append(f"{col.replace('-', ' ').title()}: {row[col]}")
        
        area_desc = f"Area: {row['area']} ."
        area_desc += " ".join(area_details)

        embedding = self.get_embedding(area_desc)


        query = """
        MERGE (a:Area {name: $area_name})
        SET a.description = $description,
            a.embedding = $embedding
        """
        # Add all available building properties dynamically
        params = {
            'area_name':row["location_y"],
            'description': area_desc,
            'embedding': embedding
        }
        
        # Add additional properties if they exist in the row
        for col in area_cols:
            if col in row and pd.notna(row[col]):
                # Create a valid property name by replacing hyphens and commas
                prop_name = col.replace('-', '_').replace(',', '').lower()
                query += f",\n        a.{prop_name} = ${prop_name}"
                params[prop_name] = str(row[col])

        self.graph.run(query, **params)

    def _parse_amenity_string(self, amenity_str) -> List[Dict]:
        """Convert string representation of amenities to List[Dict]"""
        if pd.isna(amenity_str) or not amenity_str:
            return []
        try:
            # First try using ast.literal_eval
            return ast.literal_eval(amenity_str)
        except:
            try:
                # If that fails, try json.loads
                return json.loads(amenity_str)
            except:
                print(f"Failed to parse amenity string: {amenity_str}")
                return []

    def _create_amenity_relationships(
        self, building_name: str, amenities: List[Dict],
        node_label: str, relationship_type: str
    ):
        for amenity in amenities:
            query = f"""
            MATCH (b:Building {{name: $building_name}})
            MERGE (am:{node_label} {{name: $amenity_name, vicinity: $vicinity}})
            SET am.business_status = $status,
                am.vicinity = $vicinity,
                am.latitude = $latitude,
                am.longitude = $longitude,
                am.rating = $rating,
                am.user_ratings_total = $user_ratings_total
            MERGE (b)-[:{relationship_type}]->(am)
            """
            self.graph.run(
                query,
                building_name=building_name,
                amenity_name=amenity['name'],
                vicinity=amenity['vicinity'],
                status=amenity['business_status'],
                latitude= float(amenity['latitude']),
                longitude=float(amenity['longitude']),
                rating= float(amenity['rating']),
                user_ratings_total=float(amenity['user_ratings_total']),
            )

    def query_school_analysis(self) -> List[Dict]:
        """
        Analyze areas and buildings based on school accessibility
        """
        query = """
        // Match all areas and their buildings
        MATCH (a:Area)-[:CONTAINS]->(b:Building)
        
        // Get schools for each building
        OPTIONAL MATCH (b)-[:HAS_SCHOOL]->(s:School)
        WHERE s.business_status = 'OPERATIONAL'
        
        // Get metro stations for additional context
        OPTIONAL MATCH (b)-[:HAS_METRO]->(m:MetroStation)
        WHERE m.business_status = 'OPERATIONAL'
        
        // First collect data at the building level
        WITH a, b, 
            COLLECT(DISTINCT {name: s.name, vicinity: s.vicinity}) as building_schools,
            COUNT(DISTINCT m) as building_metros
        
        // Then aggregate by area
        WITH a, 
            COUNT(DISTINCT b) as building_count,
            SUM(SIZE(building_schools)) as school_count,
            SUM(building_metros) as metro_count,
            COLLECT(DISTINCT {
                building_name: b.name,
                schools: building_schools,
                metros: building_metros
            }) as buildings
        
        // Calculate school density score
        WITH a, 
            building_count,
            school_count,
            metro_count,
            buildings,
            CASE WHEN building_count > 0 
                THEN toFloat(school_count) / building_count 
                ELSE 0 
            END as school_density
        
        RETURN a.name as area_name,
            building_count,
            school_count,
            metro_count,
            school_density,
            buildings
        ORDER BY school_density DESC
        """
            
        result = self.graph.run(query)
        return [dict(record) for record in result]
    
    def _create_building_with_amenities(self, row,row_building_info):
        # Generate embeddings for building description
        building_info_cols = [
            'about', 'in-a-nutshell', 'parking', 'elevators',
            'security,-central-air-conditioning-and-maintenance-services', 'gym',
            'lifestyle-amenities', 'concierge-and-guest-services', 'pet-friendly',
            'apartment-types', 'getting-around', 'nearby-attractions',
            'things-to-consider', 'what-type-of-apartments',
            'what-facilities-are-there'
        ]
        building_details = []
        for col in building_info_cols:
            if col in row_building_info and pd.notna(row_building_info[col]) and row_building_info[col]:
                building_details.append(f"{col.replace('-', ' ').title()}: {row_building_info[col]}")
        
        building_desc = f"Building: {row_building_info['building_name']} in {row_building_info['area_name']}. "
        building_desc += " ".join(building_details)

        embedding = self.get_embedding(building_desc)        
        # Store building with embedding
        query = """
        MATCH (a:Area {name: $area_name})
        MERGE (b:Building {name: $building_name})
        SET b.latitude = $lat,
            b.longitude = $lng,
            b.description = $description,
            b.embedding = $embedding
        """
        params = {
            'area_name': row['area_name'],
            'building_name': row['building_name'],
            'lat': float(row['latitude']),
            'lng': float(row['longitude']),
            'description': building_desc,
            'embedding': embedding
        }

        for col in building_info_cols:
            if col in row_building_info and pd.notna(row_building_info[col]):
                # Create a valid property name by replacing hyphens and commas
                prop_name = col.replace('-', '_').replace(',', '').lower()
                query += f",\n        b.{prop_name} = ${prop_name}"
                params[prop_name] = str(row_building_info[col])

        # Complete the query
        #try:
        query += "\nMERGE (a)-[:CONTAINS]->(b)"
        self.graph.run(query, **params)
        
        #except:
        #    print(query)
        #    print(params)
        #    raise
        # Create amenity relationships
        #'transport','metro_stations', 'gyms', 'spas', 'supermarkets', 'malls', 'schools','restaurant', 'hospital', 'tourist_attraction', 'park', 'worship'
        amenity_mappings = [
            ('transport', 'Transport', 'HAS_TRANSPORT'),
            ('gyms', 'Gym', 'HAS_GYM'),
            ('spas', 'SPA', 'HAS_SPA'),
            ('supermarkets', 'Supermarket', 'HAS_SUPERMARKET'),
            ('malls', 'Mall', 'HAS_MALL'),
            ('metro_stations', 'MetroStation', 'HAS_METRO'),
            ('hospital', 'Hospital', 'HAS_HOSPITAL'),
            ('schools', 'School', 'HAS_SCHOOL'),
            ('worship', 'Worship', 'HAS_WORSHIP'),
            ('restaurant', 'Restaurant', 'HAS_RESTAURANT'),
            ('park', 'Park', 'HAS_PARK'),
            ('tourist_attraction', 'TouristAttraction', 'HAS_TOURIST_ATTRACTION')
        ]

        for amenity_field, node_label, relationship_type in amenity_mappings:
            if amenity_field in row:
                amenities = self._parse_amenity_string(row[amenity_field])
                if amenities:  # Only process if we have valid amenities
                    self._create_amenity_relationships(
                        row['building_name'],
                        amenities,
                        node_label,
                        relationship_type
                    )


    def semantic_search(self, query_text: str, limit: int = 5):
        # Generate embedding for search query
        query_embedding = self.embedding_model.encode(query_text)
        
        # Search using cosine similarity in Neo4j
        result = self.graph.run("""
        MATCH (b:Building)
        WITH b, gds.similarity.cosine(b.embedding, $query_embedding) AS similarity
        WHERE similarity > 0.5
        RETURN b.name, b.description, similarity
        ORDER BY similarity DESC
        LIMIT $limit
        """, query_embedding=query_embedding.tolist(), limit=limit)
        
        return [(record["b.name"], record["b.description"], record["similarity"]) 
                for record in result]

    def hybrid_search(self, query_text: str, area_name: str = None, 
                     min_similarity: float = 0.5, limit: int = 5):
        """
        Combine semantic search with structural queries
        """
        query_embedding = self.embedding_model.encode(query_text)
        
        # Build dynamic query based on filters
        query = """
        MATCH (b:Building)
        """
        if area_name:
            query += """
            MATCH (a:Area {name: $area_name})-[:CONTAINS]->(b)
            """
        
        query += """
        WITH b, gds.similarity.cosine(b.embedding, $query_embedding) AS similarity
        WHERE similarity > $min_similarity
        RETURN b.name, b.description, similarity
        ORDER BY similarity DESC
        LIMIT $limit
        """
        
        result = self.graph.run(
            query,
            query_embedding=query_embedding.tolist(),
            area_name=area_name,
            min_similarity=min_similarity,
            limit=limit
        )
        
        return [(record["b.name"], record["b.description"], record["similarity"]) 
                for record in result]
```

```python
from langchain_community.graphs import Neo4jGraph
from langchain_community.vectorstores import Neo4jVector
from langchain_openai import OpenAIEmbeddings, ChatOpenAI
from langchain.chains import GraphCypherQAChain
from langchain.prompts import PromptTemplate
import os

class RealEstateRAG:
    def __init__(self,llm,embedding_model ,openai_api_key=None):
        # Set up OpenAI API key
        #self.api_key = openai_api_key or os.getenv("OPENAI_API_KEY")
        #if not self.api_key:
        #    raise ValueError("OpenAI API key is required")

        # Initialize Neo4j connection
        self.graph = Neo4jGraph(
            url="bolt://localhost:7687",
            username="neo4j",
            password="1234neo4j"
        )

        # Initialize embeddings
        self.embeddings = embedding_model #OpenAIEmbeddings(api_key=self.api_key)

        # Initialize vector store
        # self.vector_store = Neo4jVector.from_existing_graph(
        #     embedding=self.embeddings,
        #     url="bolt://localhost:7687",
        #     username="neo4j",
        #     password="1234neo4j",
        #     search_type="hybrid",  # Combines vector + keyword search
        #     node_label="Building",  # Main node label to search
        #     text_node_properties=["description"],  # Property containing text to embed
        #     embedding_node_property="embedding"  # Property to store embeddings
        # )
        self.llm = llm
        # Initialize LLM
        # self.llm = ChatOpenAI(
        #     temperature=0,
        #     model="gpt-4",
        #     api_key=self.api_key
        # )

        # Set up the Cypher QA Chain
        self.qa_chain = self._setup_qa_chain()

    def _setup_qa_chain(self):
        # Custom prompt template for real estate queries
        CYPHER_PROMPT = """Task: Generate a Cypher query to find real estate information based on the question.

        Database Schema:
        - (Building) nodes with properties: name, description, latitude, longitude, about, in_a_nutshell, etc.
        - (Area) nodes with properties: name, description, about_, in_a_nutshell, community_overview, etc.
        - Specific amenity nodes:
        * (Transport) with properties: name, business_status, vicinity, latitude, longitude, rating, user_ratings_total
        * (Gym) with properties: name, business_status, vicinity, latitude, longitude, rating, user_ratings_total
        * (SPA) with properties: name, business_status, vicinity, latitude, longitude, rating, user_ratings_total
        * (Supermarket) with properties: name, business_status, vicinity, latitude, longitude, rating, user_ratings_total
        * (Mall) with properties: name, business_status, vicinity, latitude, longitude, rating, user_ratings_total
        * (MetroStation) with properties: name, business_status, vicinity, latitude, longitude, rating, user_ratings_total
        * (Hospital) with properties: name, business_status, vicinity, latitude, longitude, rating, user_ratings_total
        * (School) with properties: name, business_status, vicinity, latitude, longitude, rating, user_ratings_total
        * (Worship) with properties: name, business_status, vicinity, latitude, longitude, rating, user_ratings_total
        * (Restaurant) with properties: name, business_status, vicinity, latitude, longitude, rating, user_ratings_total
        * (Park) with properties: name, business_status, vicinity, latitude, longitude, rating, user_ratings_total
        * (TouristAttraction) with properties: name, business_status, vicinity, latitude, longitude, rating, user_ratings_total
        
        - Relationships: 
        * (Area)-[:CONTAINS]->(Building)
        * (Building)-[:HAS_TRANSPORT]->(Transport)
        * (Building)-[:HAS_GYM]->(Gym)
        * (Building)-[:HAS_SPA]->(SPA)
        * (Building)-[:HAS_SUPERMARKET]->(Supermarket)
        * (Building)-[:HAS_MALL]->(Mall)
        * (Building)-[:HAS_METRO]->(MetroStation)
        * (Building)-[:HAS_HOSPITAL]->(Hospital)
        * (Building)-[:HAS_SCHOOL]->(School)
        * (Building)-[:HAS_WORSHIP]->(Worship)
        * (Building)-[:HAS_RESTAURANT]->(Restaurant)
        * (Building)-[:HAS_PARK]->(Park)
        * (Building)-[:HAS_TOURIST_ATTRACTION]->(TouristAttraction)

        Question: {question}

        Important:
        - Consider spatial relationships when relevant
        - Include ratings and review counts when available
        - Limit results to reasonable numbers (5-10)
        - Filter for operational amenities using business_status='OPERATIONAL' when appropriate
        - Only generate the Cypher query, no other text

        Cypher Query:"""

        cypher_prompt = PromptTemplate(
            template=CYPHER_PROMPT,
            input_variables=["question"]
        )

        return GraphCypherQAChain.from_llm(
            llm=self.llm,
            graph=self.graph,
            cypher_prompt=cypher_prompt,
            verbose=True,
            allow_dangerous_requests=True
        )
    def answer_query(self, query: str) -> str:
        """
        Process and answer real estate queries using graph capabilities
        """
        try:
            # Use the QA chain to get the answer directly
            # No vector store needed
            response = self.qa_chain.run(query)
            return response

        except Exception as e:
            return f"Error processing query: {str(e)}"
    def answer_query___(self, query: str) -> str:
        """
        Process and answer real estate queries using both vector and graph capabilities
        """
        try:
            # First, try to get relevant context using vector similarity
            vector_results = self.vector_store.similarity_search(
                query,
                k=3  # Get top 3 most relevant results
            )

            # Format vector results for context
            vector_context = "\n".join([doc.page_content for doc in vector_results])

            # Enhance the original query with vector context
            enhanced_query = f"""
            Consider this relevant context:
            {vector_context}

            Original question: {query}
            """

            # Use the QA chain to get the final answer
            response = self.qa_chain.run(enhanced_query)
            return response

        except Exception as e:
            return f"Error processing query: {str(e)}"
```
